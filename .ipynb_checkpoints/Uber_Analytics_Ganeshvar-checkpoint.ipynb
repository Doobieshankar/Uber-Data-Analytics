# Data Cleaning & Feature Engineering

## Convert Date Columns to Datetime

# What does `errors='coerce'` mean?

`errors='coerce'` tells Pandas:

> **"If any value cannot be converted to datetime, don't crash. Just replace it with `NaT`."**

---

### `NaT` = **Not a Time**

It is like:

* `NaN` for numbers
* `NaT` for datetime values


````python
df['tpep_pickup_datetime'] = pd.to_datetime(df['tpep_pickup_datetime'], errors='coerce')
df['tpep_dropoff_datetime'] = pd.to_datetime(df['tpep_dropoff_datetime'], errors='coerce')

print("Datetime conversion completed!")
````

````python
df.info()
````

# Create New Useful Columns

````python
# Trip duration in minutes
df['trip_duration_minutes'] = (
    (df['tpep_dropoff_datetime'] - df['tpep_pickup_datetime']).dt.total_seconds() / 60
)

# Pickup hour
df['pickup_hour'] = df['tpep_pickup_datetime'].dt.hour

# Pickup day
df['pickup_day'] = df['tpep_pickup_datetime'].dt.day

# Pickup weekday name
df['pickup_weekday'] = df['tpep_pickup_datetime'].dt.day_name()

print("New columns created successfully!")
df.head()
````

# Check Invalid or Suspicious Values

````python
print("Trips with zero or negative duration:", (df['trip_duration_minutes'] <= 0).sum())
print("Trips with zero or negative distance:", (df['trip_distance'] <= 0).sum())
print("Trips with negative fare:", (df['fare_amount'] < 0).sum())
print("Trips with negative total amount:", (df['total_amount'] < 0).sum())
````

# Optional: Remove Obvious Invalid Rows

````python
# Keep a cleaned copy
df_clean = df.copy()

# Remove invalid values
df_clean = df_clean[
    (df_clean['trip_duration_minutes'] > 0) &
    (df_clean['trip_distance'] > 0) &
    (df_clean['fare_amount'] >= 0) &
    (df_clean['total_amount'] >= 0)
]

print("Original shape:", df.shape)
print("Cleaned shape :", df_clean.shape)
````

# 🔷 1️⃣ `df_clean.sample(min(2000, len(df_clean)), random_state=42)`

## ✅ What `.sample()` Does

`.sample(n)` randomly selects **n rows** from the dataset.

---

## ✅ What is `min(2000, len(df_clean))`?

This is a safety mechanism.

### Step 1:

```python
len(df_clean)
```

Returns total number of rows.

Example:

* If dataset has 15,000 rows → `len(df_clean) = 15000`
* If dataset has 1,200 rows → `len(df_clean) = 1200`

---

### Step 2:

```python
min(2000, len(df_clean))
```

This chooses the smaller value.

### Case 1: Large dataset

```
len(df_clean) = 15000
min(2000, 15000) = 2000
```

So → sample 2000 rows.

---

### Case 2: Small dataset

```
len(df_clean) = 1200
min(2000, 1200) = 1200
```

So → sample 1200 rows (all rows).

---

## 🎯 Why do this?

If you write:

```python
df_clean.sample(2000)
```

And dataset has only 1200 rows ❌ → ERROR.

Using `min()` prevents that error.

---

## ✅ What is `random_state=42`?

It makes the random selection **reproducible**.

That means:

* Every time you run the code
* You get the **same 2000 rows**

Without `random_state`, sample changes every run.

42 is just a commonly used seed number.

---

# 🔷 2️⃣ `alpha = 0.4`

`alpha` controls **transparency**.

Range:

```
0 → Fully transparent (invisible)
1 → Fully solid
```

### `alpha = 0.4`

👉 40% opacity
👉 60% transparent

---

## 🎯 Why use transparency?

In scatter plots:

* Many points overlap
* Without transparency, dense areas look like one big dot

With `alpha=0.4`:

* Overlapping areas become darker
* You can visually detect high-density areas

Very useful for geospatial data like pickup locations.

---

# 🔷 3️⃣ `s = 10`

`s` controls **size of each point**.

* Default size is larger
* `s=10` makes points small

---

## 🎯 Why small size?

Since you're plotting up to 2000 points:

* Large size → messy
* Small size → clean map-like visualization

---

# 🔥 Quick Visual Comparison

| Parameter         | Meaning               | Effect                 |
| ----------------- | --------------------- | ---------------------- |
| `sample(2000)`    | Take 2000 random rows | Improves performance   |
| `random_state=42` | Fix random selection  | Same result every time |
| `alpha=0.4`       | Transparency          | Shows density          |
| `s=10`            | Small dot size        | Cleaner plot           |

---

# 🔥 Why Sampling is Important Here

If your dataset has 500,000 rows:

* Plotting all → Very slow
* Graph becomes unreadable
* Overplotting problem

Sampling makes:
✔ Faster
✔ Cleaner
✔ More readable
