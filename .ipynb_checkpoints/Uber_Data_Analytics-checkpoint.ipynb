# 🧹 **Section 1: Data Cleaning & Feature Engineering**

---

# **Cell 9 – Convert Datetime Columns**

```python
df['tpep_pickup_datetime'] = pd.to_datetime(df['tpep_pickup_datetime'], errors='coerce')
df['tpep_dropoff_datetime'] = pd.to_datetime(df['tpep_dropoff_datetime'], errors='coerce')

print("Datetime conversion complete.")
```

---

# **Cell 10 – Create Time & Duration Features**

```python
# Trip duration in minutes
df['trip_duration_minutes'] = (
    (df['tpep_dropoff_datetime'] - df['tpep_pickup_datetime']).dt.total_seconds() / 60
)

# Time-based features
df['pickup_hour'] = df['tpep_pickup_datetime'].dt.hour
df['pickup_day'] = df['tpep_pickup_datetime'].dt.day
df['pickup_weekday'] = df['tpep_pickup_datetime'].dt.day_name()
df['pickup_month'] = df['tpep_pickup_datetime'].dt.month

df.head()
```

---

# **Cell 11 – Create Business Metrics**

```python
# Tip percentage relative to fare amount
df['tip_percentage'] = np.where(
    df['fare_amount'] > 0,
    (df['tip_amount'] / df['fare_amount']) * 100,
    np.nan
)

# Fare per mile
df['fare_per_mile'] = np.where(
    df['trip_distance'] > 0,
    df['fare_amount'] / df['trip_distance'],
    np.nan
)

# Total per mile
df['total_per_mile'] = np.where(
    df['trip_distance'] > 0,
    df['total_amount'] / df['trip_distance'],
    np.nan
)

# Speed estimate (miles per hour)
df['avg_speed_mph'] = np.where(
    df['trip_duration_minutes'] > 0,
    df['trip_distance'] / (df['trip_duration_minutes'] / 60),
    np.nan
)

print("Business metrics created successfully.")
```

---

# **Cell 12 – Check Invalid / Suspicious Values**

```python
print("Suspicious / Invalid Value Summary")
print("-" * 50)
print("Trips with zero or negative duration :", (df['trip_duration_minutes'] <= 0).sum())
print("Trips with zero or negative distance :", (df['trip_distance'] <= 0).sum())
print("Trips with negative fare amount      :", (df['fare_amount'] < 0).sum())
print("Trips with negative total amount     :", (df['total_amount'] < 0).sum())
print("Trips with missing pickup datetime   :", df['tpep_pickup_datetime'].isnull().sum())
print("Trips with missing dropoff datetime  :", df['tpep_dropoff_datetime'].isnull().sum())
```

---

# **Cell 13 – Basic Cleaning (Create Clean Dataset)**

```python
df_clean = df.copy()

# Keep only logically valid rows
df_clean = df_clean[
    (df_clean['trip_duration_minutes'] > 0) &
    (df_clean['trip_distance'] > 0) &
    (df_clean['fare_amount'] >= 0) &
    (df_clean['total_amount'] >= 0)
].copy()

print("Original shape:", df.shape)
print("Cleaned shape :", df_clean.shape)
```

# 📦 **Section 3: IQR-Based Outlier Detection**

---

# **Cell 20 – Define IQR Helper Function**

```python
def iqr_bounds(series):
    q1 = series.quantile(0.25)
    q3 = series.quantile(0.75)
    iqr = q3 - q1
    lower = q1 - 1.5 * iqr
    upper = q3 + 1.5 * iqr
    return lower, upper
```

---

# **Cell 21 – Check Outliers in Key Columns**

```python
columns_to_check = ['trip_distance', 'fare_amount', 'trip_duration_minutes', 'total_amount']

for col in columns_to_check:
    lower, upper = iqr_bounds(df_clean[col])
    outliers = ((df_clean[col] < lower) | (df_clean[col] > upper)).sum()
    print(f"{col}:")
    print(f"  Lower Bound = {lower:.2f}")
    print(f"  Upper Bound = {upper:.2f}")
    print(f"  Outlier Count = {outliers}")
    print("-" * 40)
```

---

# **Cell 22 – Boxplots for Outlier Visualization**

```python
fig, axes = plt.subplots(2, 2, figsize=(16, 10))

sns.boxplot(y=df_clean['trip_distance'], ax=axes[0, 0])
axes[0, 0].set_title("Trip Distance")

sns.boxplot(y=df_clean['fare_amount'], ax=axes[0, 1])
axes[0, 1].set_title("Fare Amount")

sns.boxplot(y=df_clean['trip_duration_minutes'], ax=axes[1, 0])
axes[1, 0].set_title("Trip Duration (Minutes)")

sns.boxplot(y=df_clean['total_amount'], ax=axes[1, 1])
axes[1, 1].set_title("Total Amount")

plt.tight_layout()
plt.show()
```

---

# **Cell 23 – Optional: Create IQR-Capped Dataset (Without Dropping Too Much Data)**

```python
df_iqr = df_clean.copy()

for col in ['trip_distance', 'fare_amount', 'trip_duration_minutes', 'total_amount']:
    lower, upper = iqr_bounds(df_iqr[col])
    df_iqr[col] = df_iqr[col].clip(lower=lower, upper=upper)

print("IQR-capped dataset created for robust visualization.")
```

---

# **Cell 24 – Outlier Strategy Notes (Markdown Cell)**

```markdown
## Outlier Strategy
Instead of aggressively removing outliers, this notebook creates an **IQR-capped version** of selected columns.

Why this is useful:
- Preserves most of the dataset
- Reduces distortion in visualizations
- Keeps rare but valid business cases from being fully removed
```
